```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

```python
(X_train,y_train),(X_test,y_test) = keras.datasets.mnist.load_data()
```

```python
import matplotlib.pyplot as plt
plt.imshow(X_train[0])#it shows the first image 
```

```python
#since the value are from 0 to 255 in the array we need to bring them from 0 to 1
X_train = X_train/255
X_test = X_test/255
```

```python
X_train[0]
```

```python
model = Sequential()

model.add(Flatten(input_shape=(28,28)))#we hav to provide data in 784 pixels format but we have it in 28*28 format so flatten the data 
model.add(Dense(128,activation='relu'))#here u need not provide the input nodes in paramertsr
#flatten automatically converts data into 1D you have to give the input data dimension 
model.add(Dense(32,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
#in sparse u need not perform one hot endocing where as in categorical cross u have to do thata 
```

```python
history = model.fit(X_train,y_train,epochs=25,validation_split=0.2)
```

```python
y_prob = model.predict(X_test)#it gives probability for each digit like probability that the number is 1 and so on 
```

```python
y_prob
```

```python
y_pred = y_prob.argmax(axis=1)#return the val with max probabilty for each image 
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```
